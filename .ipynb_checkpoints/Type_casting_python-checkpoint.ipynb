```python
# type casting : Type casting means changing any variable type from int-float, float-str, str-int.
# examples
a=20
b=int('50') #here '50' is in str and is converted to int
c=a+b
print(c)
```

```python
#example
a=str(50) #here 50 is in int and converted it into str
b='20'
c=a+b
print(c)
```

```python
#examples
a=float(20)  #converting int to float
b=15
c=a+b
print(c)
```

```python
#examples
a=22.55
b=97.77
c=a+b
print(c)
```

```python
#examples
a=int(22.55) #converting float to int
b=int(97.77) #converting float to int
c=a+b
print(c)
```

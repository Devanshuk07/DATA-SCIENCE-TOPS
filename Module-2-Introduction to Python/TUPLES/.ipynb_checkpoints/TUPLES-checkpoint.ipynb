```python
# TUPLES ARE builtin data type which is immutable that cannot be changed and works as list
# examples :
# the only diffence between list and tuples is list=[] and tuple=() and tuple is immutable which cannot change

tup=(2,1,3,1,3,4,5)
# tup(0)=5   # cannot assign value in tuples
print(type(tup))
```

```python
# Slicing in turple is same as list

tup=(2,1,3,1,3,4,5)
print(tup[0:5])
```

```python
#index in turple

tup=(2,1,3,1,3,4,5)
print(tup.index(3))  # means (3) is in 2nd place.This is called indexing
```

```python
#count in turple

tup=(2,1,3,1,3,4,5,8,9,0,3,7,3,8,9,2)
print(tup.count(3))     #   Means 3 is 4 times in turple list.
```

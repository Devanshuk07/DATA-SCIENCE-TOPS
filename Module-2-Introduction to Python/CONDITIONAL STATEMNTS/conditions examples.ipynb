```python
a=input("Enter Which Day Is Today :")
if a=="monday":
    print("Today Is NOT Monday")
elif a=="tuesday":
    print("Today Is NOT Tuesday")
elif a=="wednesday":
    print("Yes ! Today Is Wednesday")
elif a=="thursday":
    print("Today Is NOT Thursday")
elif a=="friday":
    print("Today Is NOT Friday")
elif a=="saturday":
    print("Today Is NOT Saturday")
elif a=="sunday":
    print("Today Is NOT Sunday")
else:
    print("PLEASE Enter Name OF WEEK")
```

```python
name=input("Enter Any Alphabet :").lower()   #.lower() always take small letter from user .upper()=for upper case
char=['a','e','i','o','u']
if name in char:
    print("The Alphabet Is a VOWEL")
else:
    print("The Alphabet Is Consonent")
```

```python
print("1=ADDITION 2=SUBSTRACTION 3=DIVISION 4=MULTIPLICATION")
a=int(input("Enter The Value For A :"))
b=int(input("Enter The Value For B :"))
c=int(input("Enter The Digit(Formula) You Want to perform :"))
if c==1:
    d=a+b
    print(d)
elif c==2:
    d=a-b
    print(d)
elif c==3:
    if b==0:
        print("Enter the number other then Zero")
    else:
        d=a/b
        print(d)
elif c==4:
    d=a*b
    print(d)
```

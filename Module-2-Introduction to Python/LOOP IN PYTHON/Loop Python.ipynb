```python
'''FOR LOOP:
FOR LOOP IS EXECUTED WHEN CONDITION IS TRUE
FORMAT:-

FOR (VARIABLE) IN RANGE(FROM,TILL)  #RANGE IS FOR NUMBERS ONLY
    PRINT(VARIABLE)
'''
```

```python
for a in range(1,100):  #range is for numbers only
    print(a)
```

```python
for a in range(1,20):   
    print("MY NAME IS KD",a)    #will print it 19 times, if you want to print it 20 times  range(1,21)
```

```python
#We can define all employees in list and in loop  print in a perticular format
employees=['devanshu','dev','jinal','rutu','divya','dhruv','rajesh','minal']
for emp in employees:   #emp taken as variable
    print(emp)
```

```python
#write a program to print even numbers between range
for i in range(1,20):
    if i%2==0:
        print(i)
```

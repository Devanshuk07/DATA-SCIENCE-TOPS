```python
i=0       # I =0 (INITIALLY)
while i<=10:   #CONDITION GIVEN I=0 :0<=10
    print(i)   #PRINT 0
    i+=1     #THEN I=I+1  I=1 THEN AGAIN LOOP WILL CONTINUE FROM WHILE
print("LOOP ENDS AT VALUE OF I :",i)     #AT LAST the value of i will be 11 so it will not fulfill the condition so loop will be TERMINATED
```

```python
#Sum to find n natural numbers
a=int(input("Enter FIRST Natural Numbers You Want sum :"))
b=int(input("Enter LAST Natural Numbers You Want sum :"))
sum=0
i=a     # i= user input 1st value lets say 2 and b=3
while(a<=i<=b):  # condition is 2<=i<=3
    sum=sum+i    #0=0+2=2
    i=i+1
print("SUM OF NATURAL NUMBERS ARE :",sum)
```

```python
i=10
while i>=1:   # THIS IS STOPPING CONDITION
    print(i)
    i-=1    # WE STARTED FROM 10 SO 10-1=9 LIKE THIS LOOP WILL CONTINUE 
print("LOOP ENDED")
```

```python
#fibonacci series
num=int(input("ENTER THE NUMBER :"))
a,b=0,1    #INITIAL VALUE FOR A,B
i=0
while i<num:  #I IS LESS THAN NUM CONDITION
    print(a,end=" ")  # A=0 
    a,b=b,a+b    # A=B AND B=A+B LOGIC FOR SERIES
    i+=1
```

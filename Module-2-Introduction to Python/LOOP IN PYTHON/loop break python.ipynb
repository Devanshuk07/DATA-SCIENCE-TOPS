```python
'''BREAK AND CONTINUE IN LOOP'''
'''
BREAK IS LIKE STATEMENT IF YOU WANT TO GIVE CONDITION IF IS USED(NO CONDITIONS IN BREAK)
CONTINUE IS LIKE STATEMENT IF YOU WANT TO GIVE CONDITION IF IS USED(NO CONDITIONS IN CONTINUE) '''
```

```python
#BREAK IS USED TO BREAK THE LOOP AND TERMINATE THE LOOP
for i in range(1,99):
    if i==51:      # THIS IS THE CONDITION THAT YOU GIVE TO BREAK THE LOOP
        break      # IT WILL BREAK THE LOOP AS THE CONDITION YOU HAS GIVEN
    print(i)
```

```python
# CONTINUE IS USED TO SKIP SOME VALUES AND THEN CONTINUING THE REST OF LOOP
for i in range(1,51):
    if i==5 or i==10 or i==20 or i==30 or i==40 or i==50:  #THIS ALL VALUES WILL NOT BE PRINTED AND THE REST WILL BE CONTINUED
        continue
    print(i)
```

```python
#compound interest
#CI= P*(1+R/100)pow(T)
P=float(input('ENTER PRINCIPAL :'))
R=float(input('ENTER RATE :'))
T=float(input('ENTER TIME :'))
total= P*(pow((1+R/100),T))
print("AMOUNT(PRINCIPAL+INTEREST) :",total)
CI=P*(pow((1+R/100),T))-P
print("INTEREST :",CI)
```

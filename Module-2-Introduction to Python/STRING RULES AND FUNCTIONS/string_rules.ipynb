```python
#upper case string
name="Hi My Name Is Devanshu"
result=name.upper()  #variable_name.upper() for full string upper case
print(result)
```

```python
#lower case string
name="Hi My Name Is Devanshu"
result=name.lower()  #variable_name.lower() for full string upper case
print(result)
```

```python
#endswith

str="i am studing in apnacollege"
print(str.endswith("ege"))
```

```python
#capitalize

str="i am studing in apnacollege"
print(str.capitalize())   #capitalize only first Charactor
print(str)    #dont effect the original string
```

```python
#replace

str="i am studing in apnacollege"
print(str.replace("a","o"))    #replace all the a to o
```

```python
#find

str="i am studing in apnacollege"
print(str.find("n"))    #counts the letters from left to right starting from 0 and indexing the n letter value=10
print(str.find("co"))   #co in college = 20 value
```

```python
#count

str="i am studing in apnacollege"
print(str.count("i"))      # countsi in full string and give how much time it comes
print(str.count("studing")) #count studing in full string and return how much time it comes
```

```python
#lower() // upper() 

name=input("Enter Any Alphabet :").lower()   #.lower() always take small letter from user .upper()=for upper case
char=['a','e','i','o','u']
if name in char:
    print("The Alphabet Is a VOWEL")
else:
    print("The Alphabet Is Consonent")
```

```python
# append means adding to the list syntax= variable.append(value you want to add)
list=[2,1,3]
list.append(8)  #append means adding to the list.
print(list)
```

```python
# sort function used to sort the list in acending order. syntax=variable.sort().
list=[2,1,3]
list.sort()
print(list)
```

```python
# sori(reverse=true) used to sort in decending order syntax=variable.sort(reverse=True(T always capital))
list=[2,1,3]
list.sort(reverse=True)
print(list)
```

```python
# reverse
list=['banana','lichi','apple']
list.sort(reverse=True)
print(list)
```

```python
#direct sort
list=['banana','lichi','apple']
list.sort()
print(list)
```

```python
#reverse() will reverse the original string. syntax list.reverse().
list=['banana','lichi','apple']
# list.append("mango")
list.reverse()   
list.append("mango")
print(list)
```

```python
# list.insert() used to insert the value syntax=list.insert(place,value).
list=[2,1,3]
list.insert(1,5)  # it means in 1st place value 5 is inserted and at 0th 2 is there already.
print(list)
```

```python
# REMOVE used to remove a perticular value. Syntax=list.remove(value)
list=[2,1,3,7,8,9,10,11]
list.remove(8)   # the value of 8 will be removed
print(list)
```

```python
#POP is used to remove a perticular value from its place.syntax=list.pop(place_of_value_to_remove)
list=[2,1,3,7,8,9,10,11]
list.pop(5)   # at 5th place value is 8 and it will be removed
print(list)
```
